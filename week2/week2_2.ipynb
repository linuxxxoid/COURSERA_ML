```python
from numpy.random import exponential
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig, axes = plt.subplots()
axes.plot(exponential(5, 20))
axes.set_title('Graphic')
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 3, figsize = (10, 5))
for row, row_axes in enumerate(axes):
    for column, ax in enumerate(row_axes):
        ax.plot(exponential(column, 20))
        ax.set_title('Holst column {} string {}'.format(column + 1, row + 1))
fig.tight_layout()
```

```python
fig.savefig('week2_2_graphics.png')
```
