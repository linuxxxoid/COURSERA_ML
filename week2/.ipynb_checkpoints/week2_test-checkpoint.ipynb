Adding needed libraries 

```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import plotly.offline as offline
offline.init_notebook_mode(connected = True)
from plotly.graph_objs import *
import plotly.tools as tls
```

Import dataset

```python
ds = pd.read_csv('fifa/CompleteDataset.csv', header = 0)
```

```python
ds.head()
```

Let's index the dataset starting from 1.

```python
ds['Index'] = ds['Unnamed: 0'] + 1
```

Take the columns we will use.

```python
ds = ds[['Index', 'Name', 'Age',
         'Nationality', 'Overall', 'Potential',
         'Club', 'Value', 'Wage', 'Preferred Positions']]
```

```python
ds.head()
```

Getting some info about dataset.

```python
ds.info()
```

Values in columns 'Value' and 'Wage' don't contain numeric but objects, thus we need to preprocess the data to make it usable for future working.

Function 'converter' converts values from a column of type object to numbers (float).
To do this, create two columns of float type 'Value_numbers', 'Wage_numbers', which will correspond to the columns 'Value', 'Wage' of type object.

```python
def converter(value):
    if value[-1] == 'K':
        return float(value[1:-1]) * 1000
    elif value[-1] == 'M':
        return float(value[1:-1]) * 1000000
    else:
        return float(value[1:])
```

```python
ds['Value_numbers'] = ds['Value'].apply(lambda x: converter(x))
ds['Wage_numbers'] = ds['Wage'].apply(lambda x: converter(x))
```

```python
ds.head()
```

Also we can see that in columns 'Preferred positions' we have more than one position in line. So we should fix it by splitting the string. I will take into account only the first position in the string as main feature.

```python
ds['Position'] = ds['Preferred Positions'].str.split().str[0]
ds.head(10)
```

The 'Preferred Position' histogram of the players.

It will indicate the dominant positions and distribution characteristics.

Here we can see that most of the players prefer CB, GK, ST, CDM positions.

```python
plt.figure(figsize = (10, 5))
plt.title('Grouping players by Preferred Position', fontsize = 20, fontweight='bold')
plt.xlabel('Position', fontsize = 15)
plt.ylabel('Number of players', fontsize = 15)
sns.countplot(x = 'Position', data = ds);
plt.show()
```

The 'Age' histogram of the players.

It will indicate the dominant age and distribution characteristics.

The dominant age is 25.

```python
plt.figure(figsize = (10, 5))
plt.title('Grouping players by Age', fontsize = 20, fontweight = 'bold')
plt.xlabel('Number of players', fontsize = 15)
plt.ylabel('Players Age', fontsize = 15)
sns.countplot(x = 'Age', data = ds);
plt.show()
```

Comparison of the dependence of the wage on the age of the player.

We should see the dependence of the wage on the age of the player. 

This graph shows that most of all ages receive the same wage (about 100000).

```python
plt.figure(figsize = (10, 5))
plt.title('Players wage according to their Age', fontsize = 20, fontweight = 'bold')
plt.xlabel('Age', fontsize = 15)
plt.ylabel('Wage_numbers', fontsize = 15)

plt.scatter(x = ds['Age'].values, y = ds['Wage_numbers'].values, color = 'green')
plt.show()
```

Let's show the most valuable football players.

Here we can see how much the best 100 players were bought by the club.

```python
top100players = ds.groupby('Name')['Value_numbers'].max().sort_values(ascending = False).head(100)

trace = Scatter(
    x = ds.Index,
    y = top100players,
    text = ds.Name,
    marker = dict(color = 'black'),
    mode = 'markers'
)

data = [trace]
layout = Layout(
    title = 'Top-100 players value',
    xaxis = dict(title = 'Array'),
    yaxis = dict(title = 'Total Value')
)

fig = Figure(data = data, layout = layout)
offline.iplot(fig)
```

Find top-10 footbal clubs paying maximum wage.

Here we can see the richest clubs, which were determined by calculating the wage of their players.

```python
top10team = ds.groupby('Club')['Wage_numbers'].mean().sort_values(ascending = False).head(10)

trace = Bar(
    x = ds.Index,
    y = top10team,
    text = ds.Club,
    marker = dict(color = 'blue'))

data = [trace]
layout = Layout(barmode = 'group',
                  title = 'Top 10 teams paying maximum wage',
                  xaxis = dict(title = 'Array'),
                  yaxis = dict(title = 'Total Wage'))

fig = Figure(data = data, layout = layout)
offline.iplot(fig)
```

Let's count players from the same country and show it in the world map. 

The brightest parts of the map indicate that most of the players were born here.

```python
players_same_country = ds.groupby('Nationality').Name.count().sort_values(ascending = False)

trace = Choropleth(
    locationmode = 'country names',
    locations = players_same_country.index,
    text = players_same_country.index,
    z = players_same_country.values,
    colorscale= 'speed'
)

data = [trace]

layout = Layout(
    title = 'Number of players from over the world'
)

fig = Figure(data = data, layout = layout)
offline.iplot(fig)
```
