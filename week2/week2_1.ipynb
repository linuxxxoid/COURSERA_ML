```python
data1 = exponential(5, 20)
data2= exponential(6, 20)
```

```python
plt.plot(data1)
```

```python
plt.scatter(range(len(data2)), data2)
```

```python
plt.plot(data1)
plt.scatter(range(len(data2)), data2)
```

```python
plt.plot(data1, label = 'First')
plt.scatter(range(len(data2)), data2, label = "Second", color = 'red')
plt.title('Results', fontdict = {'fontsize': 20})
plt.xlabel('Attempt number')
plt.ylabel('Result')
plt.legend()
plt.xticks(range(0, 20, 4))
```
